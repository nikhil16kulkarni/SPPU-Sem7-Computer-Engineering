### Scatter Plots

```python
iris_feat = df.iloc[:,:-1]
iris_species = df.iloc[:,-1]

iris_feat.columns
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x = iris_feat['SepalLengthCm'],y = [iris_feat['PetalLengthCm']],alpha=1)

plt.title('Scatter Plot')
plt.xlabel('sepal_length')
plt.ylabel('petal_length')

plt.show()
```

```python
import seaborn as sns
```

```python
colors = {'Iris-setosa':'r', 'Iris-virginica':'g', 'Iris-versicolor':'b'}

fig, ax = plt.subplots()

for i in range(len(iris_feat['SepalLengthCm'])):
    ax.scatter(iris_feat['SepalLengthCm'][i], iris_feat['SepalWidthCm'][i],color=colors[iris_species[i]])
    
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')

plt.show()
```

### Heatmap

```python
corr = df.corr()

fig, ax = plt.subplots()
img = ax.imshow(corr.values,cmap = "magma_r")

ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)
plt.setp(ax.get_xticklabels(), rotation=30, ha="right",
         rotation_mode="anchor")

for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        if corr.iloc[i, j]<0:
            text = ax.text(j, i, np.around(corr.iloc[i, j], decimals=2),
                       ha="center", va="center", color="black")
        else:
            text = ax.text(j, i, np.around(corr.iloc[i, j], decimals=2),
                       ha="center", va="center", color="white")
```

### Box Plot

```python
sepal_length = iris_feat['SepalLengthCm']
petal_length = iris_feat['PetalLengthCm']
petal_width = iris_feat['PetalWidthCm']
sepal_width = iris_feat['SepalWidthCm']

data = [sepal_length , petal_length , petal_width , sepal_width]
fig1, ax1 = plt.subplots()
ax1.set_title('Basic Plot')
ax1.boxplot(data)
plt.show()
```
